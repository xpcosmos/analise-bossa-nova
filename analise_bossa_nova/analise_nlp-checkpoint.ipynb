# Bossa nova: Um advento da música brasileira

## Proposta

A bossa nova é sem vias de dúvidas um dos momentos mais importantes para a cultura social brasileira. O ritmo surgiu logo após a segunda guerra mundial, por volta de 1958. Naquela época era visado entre músicos de classe média alta a internacionalização da música brasileira. Por meio de um canto mais falado e com menos vocais proeminentes, nasceu a **bossa nova**.

Com o acontecimento do fato histórico da ditadura militar, notou-se haver uma maior presença de temas sociais nas músicas, o que por sua vez significou a extinção  por volta de 1966, extinção essa teórica, mas não prática. A musica que procedeu a bossa nova foi a **MPB**. Esse gênero deixou ainda mais claro a instisfação da classe artística com o regime militar.

Visando entender mais sobre esse ritmo e sobre a importância dele para o cenário musical brasileiro, utlizaremos uma base de dados contendo músicas de bossa nova. Vamos extrair a _data_ utilizando técnicas de NLTK _(processamento de linguagem natural)_ e tentaremos trazer os insights através de uma biblioteca de _WordCloud_.

## Limpeza de dados

Vamos criar um print() com todas as primeias 10 palavras e suas respectivas frequência dentro da matriz:

```python
top_dict= {}

a = 0
for i in matriz_data_pt.columns:
    top = matriz_data_pt[i].sort_values(ascending=False).head(30)
    top_dict[i] = list(zip(top.index, top.values))
 

for artista, palavra in top_dict.items():
    print(artista + '\n')
    print(', '.join([str(i) for i in palavra[:14]]))
    print('\n _____________________________________________________________________________')
    a = a+1
    if a == 3:
        break
        
# Reduzi a quantidade de dados imprimidos para facilitar a leitura do notebook.

    
```

Vamos importar a função 'Counter' para nos ajudar com a contagem das palavras:

```python
from collections import Counter
```

É interessante que criemos um loop para ver a quantidade de palavras mais comuns dentre as dos artistas:

```python
palavras_comuns = []
for artista in matriz_data_pt.columns:
    top = [palavra for (palavra, contagem) in top_dict[artista]]
    for t in top:
        palavras_comuns.append(t)
palavras_comuns[:10]
```

```python
Counter(palavras_comuns).most_common()[:10]
```

Vamos eliminar as palavras mais comuns: 

Eu optei por eliminar as que aparecem em até 40 artistas em comum, mas você pode verificar qual se encaixa melhor no seu DataFrame. Vai depende do volume de dados e o tipo de análise que está fazendo!

Note que optei por não eliminar 'Amor', pois acho que podemos tirar conclusões do uso dessa palavra.

```python
stopwords = nltk.corpus.stopwords.words('portuguese')

add_stopword = [palavra for palavra, contagem in Counter(palavras_comuns).most_common() if contagem > 40]
add_stopword.remove('amor')
```

### Visualização de dados

Vamos dar uma verificada nos resultados que podemos encontrar com as palavras que estão em nossos tokens. Para isso irei utilizar as bibliotecas matplotlib e WordCloud, PIL, numpy e scipy para trabalharmos com imagens e aplicarmos máscaras interessantes.

```python
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import gaussian_gradient_magnitude
import numpy as np
from wordcloud import WordCloud, ImageColorGenerator
```

Vou mesclar todas as músicas pois o que nos interessa aqui é o gênero bossa nova como um todo.

```python
texto = songs_lyrics_artista.lyrics.values
texto[0]
```

```python
def word_cloud_visualization(stop_words, dados, nome_arquivo, qnt_palavras):
    

    
    
    mask_color = np.array(Image.open(r"C:\Users\mikei\OneDrive\Documentos\analise_bossa_nova\mascaras\mask-image.png"))
    mask_color = mask_color[::3, ::3]
    mask_image = mask_color.copy()
    mask_image[mask_image.sum(axis=2) == 0] = 255
    

    edges = np.mean([gaussian_gradient_magnitude(mask_color[:, :, i]/255., 2) for i in range(3)], axis=0)
    mask_image[edges > .08] = 255
    
    wc = WordCloud(max_words=qnt_palavras, stopwords = stop_words, background_color ='white',
                   max_font_size=150, random_state = 42, width=1200, height=600, mask=mask_image)

    


    wc.generate(str(texto))
    image_colors = ImageColorGenerator(mask_color)
    wc.recolor(color_func=image_colors)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    
    plt.figure(dpi = 300)
    pasta = r"C:\Users\mikei\OneDrive\Documentos\analise_bossa_nova\word_cloud\\"
    wc.to_file(pasta[:-1] + nome_arquivo + ".png")
    plt.show()
```

```python
word_cloud_visualization(stop_words, texto, 'analise_0', 100)
```
